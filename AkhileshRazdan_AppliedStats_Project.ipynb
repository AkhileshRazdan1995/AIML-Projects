#### <span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#ff7b56;"> Question 1: 

Please refer the table below to answer below questions:

| Planned to purchase Product A  | Actually placed and order for Product A - Yes  |  Actually placed and order for Product A - No |  Total |
|---|---|---|---|
| Yes  | 400  |  100 | 500  |
| No  | 200  | 1300  | 1500 |
| Total  | 600  | 1400  | 2000  | 
    
#### A) Refer to the above table and find the joint probability of the people who planned to purchase and actually 
#### placed an order.

#### B) Refer to the above table and find the joint probability of the people who planned to purchase and actually 
#### placed an order, given that people planned to purchase.

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#c3d500;"> Solution A :

Pre work : Creating the data frame as given in above table

Steps for A:
* Take the total number of people planned to purchase and actually placed an order. (number of Favorable Outcomes of an event)
* Take the total number of people in the data. ( The total number of outcomes)

$$Probability = \frac{number\;of\;Favorable\;Outcomes\;of\;an\;event}{total\;number\;of\;outcomes} $$

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#c3d500;"> Code Solution:

```python
# Importing required pakages

import numpy             as np
import pandas            as pd
import scipy.stats       as stats
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind
from statsmodels.stats.proportion import proportions_ztest
from scipy.stats import f_oneway


%matplotlib inline 

# Creating dataframe

df = pd.DataFrame({'Actually placed and order for Product A - Yes': [400,200,600],
                   'Actually placed and order for Product A - No': [100,1300,1400],
                   'Total': [500, 1500, 2000]},
                  index=['Yes', 'No', 'Total'])

df
```

```python
# NumberofFavorableOutcomes is the number of people who planned and actually placed an order for Product A (number of Favorable Outcomes of an event)
NumberofFavorableOutcomes = df.at['Yes','Actually placed and order for Product A - Yes']

# TotalNumberOfOutcomes is the total number of people in the data ( The total number of outcomes)
TotalNumberOfOutcomes  = df.at['Total','Total']

# JointProbilibityOfEvent is the Joint probability of the people who planned to purchase and actually placed an order
JointProbilibityOfEvent = round(NumberofFavorableOutcomes / TotalNumberOfOutcomes,5) 

print('\033[1m' + 'Joint probability of the people who planned to purchase and actually placed an order is %1.5f' % JointProbilibityOfEvent  + '\033[0m')
```

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#c3d500;"> Solution B :

Pre work : Creating the data frame as given in above table [Alreday Done above]

Steps for B:
* For B we have take into account the Conditional Probability.
 
$$Probability = \frac{P(A \quad and \quad B)}{P(B)} $$

$$P(A\mid B) = \frac{P(A \quad and \quad B)}{P(B)}$$

where P(A and B) = Joint probability of A and B
*     P(A)       = Marginal probability of A
*     P(B)       = Marginal probability of B


Steps:   
* Take the joint probability of people planned to purchase and actually placed an order. 
* Take the Marginal probability of B. 
* Divide joint probability of people planned to purchase and actually placed an order by the Marginal probability of B


    
<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#c3d500;"> Code Solution:

```python
#Finding joint Probability of people planned to purchase the Product and actually placed an order

# NumberofFavorableOutcomesForA is the number of people who planned and actually placed an order for Product A (number of Favorable Outcomes of an event)
NumberofFavorableOutcomesForA = df.at['Yes','Actually placed and order for Product A - Yes']

# TotalNumberOfOutcomes is the total number of people in the data ( The total number of outcomes)
TotalNumberOfOutcomes  = df.at['Total','Total']

# JointProbilibityOfEventforA is the Joint probability of the people who planned to purchase and actually placed an order
JointProbilibityOfEventForA = round(NumberofFavorableOutcomesForA / TotalNumberOfOutcomes,5) 

print('Joint probability of the people who planned to purchase and actually placed an order. is %1.5f' % JointProbilibityOfEventForA)

#Finding Marginal probability of people planned to purchase Product
# NumberofFavorableOutcomesForB is the number of people who planned to placed an order for Product A 
NumberofFavorableOutcomesForB=df.at['Yes','Total']

# ProbilibityOfEventForB is the probability of the people who planned to placed an order for Product A
ProbilibityOfEventForB= round(NumberofFavorableOutcomesForB/TotalNumberOfOutcomes,5)

print('Probability of the people who planned to purchase the product is %1.5f' % ProbilibityOfEventForB)
```

```python
ConditionalProbability= round(JointProbilibityOfEventForA/ProbilibityOfEventForB, 5)

print('\033[1m' + 'Probability of A given B = Joint Probability Planned to purchase and actually placed an order/Probability Planned to purchase =  %1.5f' % ConditionalProbability +'\033[0m' )
```

#### <span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#ff7b56;"> Question 2: 
An electrical manufacturing company conducts quality checks at specified periods on the 
products it manufactures. Historically, the failure rate for the manufactured item is 5%. Suppose a random 
sample of 10 manufactured items is selected. Answer the following questions.
#### A. Probability that none of the items are defective?
#### B. Probability that exactly one of the items is defective?
#### C. Probability that two or fewer of the items are defective?
#### D. Probability that three or more of the items are defective 

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#c3d500;"> Solution A and B :

We will be using the Binomial Distribution function present in stats pakage.

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#c3d500;"> Solution C and D :

We will be using the cumulative probability function present in stats pakage.

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#c3d500;">  Code Solution:

```python
# failure rate for the item that the company manufactures is 5%
FailureRate   =  0.05 

# sample size of the test conducted
SampleSize   =  10 

# An array of quantiles
RangeForSample   =  np.arange(0,SampleSize+1) 

# Binominal Probability of defective items for the samples size
Binomial = stats.binom.pmf(RangeForSample,SampleSize,FailureRate)

print(Binomial)

print('\033[1m' +'A) Probability None of the items are defective is %1.5f' %Binomial[0]+ '\033[0m')

print('\033[1m' +'B) Probability Exactly one of the items are defective is %1.5f' %Binomial[1] +'\033[0m')
```

```python
# Cumulative Binomial Probability of defective items for the samples size
CumulativeBinomial = stats.binom.cdf(RangeForSample,SampleSize,FailureRate)

print(CumulativeBinomial)

print('\033[1m' +'C) Probability that two or fewer of the items are defective is %1.5f' %CumulativeBinomial[2]+ '\033[0m')

print('\033[1m' +'D) Probability that three or more of the items are defective is %1.5f' %(1-CumulativeBinomial[2]) +'\033[0m')
```

#### <span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#ff7b56;"> Question 3: 
A car salesman sells on an average 3 cars per week. 
##### A. Probability that in a given week he will sell some cars.
##### B. Probability that in a given week he will sell 2 or more but less than 5 cars.
##### C. Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week.

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#c3d500;"> Solution A and B :

We will be using the Possion cumulative Distribution probability present in stats pakage.
Note Sampling in possion distribution are around 5 * Rate or 7 * Rate for get sufficent probability

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#c3d500;"> Solution C :

We will be using the pyplot function present in matplotlib pakage and Possion Distribution Function.

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#c3d500;">  Code Solution:

```python
# rate of sale of cars per week
RateofSale=3

# Sampling size as per rate of sale of cars per week
NumberOfSamples= np.arange(0,((RateofSale*5)+1))

#Cumulative Possion Probability is required
CumulativePossionProbability = stats.poisson.cdf(NumberOfSamples,RateofSale)
print(CumulativePossionProbability)

# ProbabilityOfSale probability that the salesman sells some cars
ProbabilityOfSale= (1 - CumulativePossionProbability[0])
print('\033[1m' + 'A) Probability that the agent sells some cars is %1.5f' % ProbabilityOfSale + '\033[0m')

# ProbabilityOfSale probability that the salesman sells Less than 5 cars but more than 1 car
# Let P1 be the probability that the salesman sells more than 4 cars
 # P(X >=4)
ProbabilityOfSaleForMoreThan4Cars = CumulativePossionProbability[4]

# Let P2 be the probability that the salesman sells more than 1 cars
# P(X >=1)
ProbabilityOfSaleForMoreThan1Car = CumulativePossionProbability[1] 

# ProbabilityOfSale probability that the agent sells some cars
# ProbabilityOfSale that the agent will sell between 2 and 5 cars
ProbabilityOfSale = ProbabilityOfSaleForMoreThan4Cars - ProbabilityOfSaleForMoreThan1Car 
print('\033[1m' + 'B) Probability that the agent sells between 2 and 5 cars is %1.5f' % ProbabilityOfSale + '\033[0m')

#Possion Distribution Function for Graph ploting
Poisson = stats.poisson.pmf(NumberOfSamples,RateofSale)
print('\033[1m' + 'C)Ploting the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week' + '\033[0m')
plt.plot(NumberOfSamples,Poisson, 'o-', color="Red")
plt.title('Poisson')
plt.xlabel('Number of cars sold per week')
plt.ylabel('Cumulative probability of cars sold per-week')
plt.grid()
plt.show()
```

86.8%. Suppose that you place order with the bot and two friends of yours independently place orders with 
the same bot. Answer the following questions.
##### A. What is the probability that all three orders will be recognised correctly?
##### B. What is the probability that none of the three orders will be recognised correctly?
##### C. What is the probability that at least two of the three orders will be recognised correctly?

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#c3d500;"> Solution A , B and C :

We will be using the Binomial Distribution function present in stats pakage.

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#c3d500;">  Code Solution:

```python
# failure rate for the item that the company manufactures is 13.2%
SuccessRate = 0.868
FailureRate = (1-SuccessRate)

# sample size of the test conducted
SampleSize   =  3

# An array of quantiles
RangeForSample   =  np.arange(0,SampleSize+1) 

# Binominal Probability of defective items for the samples size
Binomial = stats.binom.pmf(RangeForSample,SampleSize,FailureRate)

print(Binomial)

print('\033[1m' +'A) Probability that all three orders will be recognised correctly is %1.5f' %Binomial[0]+ '\033[0m')

print('\033[1m' +'B) Probability that none of the three orders will be recognised correctly is %1.5f' %Binomial[3] +'\033[0m')

print('\033[1m' +'C) Probability that at least two of the three orders will be recognised correctly is %1.5f' %(Binomial[0]+Binomial[1]) +'\033[0m')
```

##### A. What is the percentage of students who score more than 80.
##### B. What is the percentage of students who score less than 50.
##### C. What should be the distinction mark if the highest 10% of students are to be awarded distinction

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#c3d500;"> Solution A , B and C :

We will be using the Normal Distubution Cure with Z test to get the required data. 

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#c3d500;">  Code Solution:

```python
def ZtestValue(x,MeanOfTheScore,StandardDeviation):
    z=((x-MeanOfTheScore)/StandardDeviation)
    return z

def reverseofZtest(z,MeanOfTheScore,StandardDeviation):

    x=((z*StandardDeviation)+MeanOfTheScore)
    return x

MeanOfTheScore=60
StandardDeviation=12
X1=80
X2=50

#Finding percentage of students who score more than 80
Zvalue1= ZtestValue(X1,MeanOfTheScore,StandardDeviation)
print('Zvalue for students who scored more that 80 is %1.5f' %Zvalue1 +' From Ztable %1.5f' %Zvalue1+ 'percentage of area under normal distribution curve is 95.25 %')
print('\033[1m' +'A.) Percentage of students who score more than 80  is %1.5f' %(100-95.25)+'%'+'\033[0m' )

#Finding percentage of students who score less than 50
Zvalue2= ZtestValue(X2,MeanOfTheScore,StandardDeviation)
print('Zvalue for students who scored less that 50 is %1.5f' %Zvalue2 +' From Ztable %1.5f' %Zvalue2+ 'percentage of area under normal distribution curve on positive side is 79.67 %')
print('\033[1m' +'B.) Percentage of students who score less than 50 is %1.5f' %(100-79.67)+'%'+'\033[0m' )

#Finding distinction mark if the highest 10% of students are to be awarded distinction 
# reversing the process top 10 % is need there for Zvalue at positive 90 has to found which is equal to 1.29
Zvalue3=1.29
Xvalue= reverseofZtest(Zvalue3,MeanOfTheScore,StandardDeviation)

print('For Zvalue %1.5f' %Zvalue3+' X is equal to %1.5f' %Xvalue +' marks')
print('\033[1m' +'C.) The distinction mark if the highest 10 percent of students are to be awarded distinction is %1.5f' %Xvalue +' and above' +'\033[0m' )
```

Test for significance and conclusion<br>
<span style="font-family: Arial; font-weight:bold;font-size:0.5.5em;color:#c3d500;"> Solution :
    
**Alternate hypothesis (Ha)** : There is significant difference between the two means<br> Mean funding of Operating Companpies != Mean funding of Operating Companpies <br>
**Null hypothesis (Ho)** : There is no difference between the two means<br>Mean funding of Operating Companpies = Mean funding of Operating Companpies <br>



```python
#for sol we are using t test

FundsInOperatingCompanies = DataFrameForWareHouseWithoutNAValues.loc[DataFrameForWareHouseWithoutNAValues.OperatingState == 'Operating', 'FundsInMillion']
FundsInClosedCompanies = DataFrameForWareHouseWithoutNAValues.loc[DataFrameForWareHouseWithoutNAValues.OperatingState =='Closed', 'FundsInMillion']
# Let's consider a accepectable error level of 5%
alpha = 0.05  


test_statistic, p_value = ttest_ind(FundsInOperatingCompanies, FundsInClosedCompanies)

print('test_statistic Value : %1.3f' %test_statistic)
print('p_value  : %1.3f' %p_value)
```

<span style="font-family: Arial; font-weight:bold;font-size:0.5.5em;color:#c3d500;"> Answer :<br>
** As the P_value is significantly greater than alpha (accepectable error i.e. 0.193>0.05) we fail to reject the Null hypothesis**<br>
    
### Conclusion:<br>
 
From the tests above we found no evidence to state that the companies that have more funding are likely to suceed more or vice-versa
